### Task 1: Creating a function that performs basic operation in a dataframe

```python
try:
  with open('/USA_cars_datasets.csv') as file:
    #set the column to vin since this is the primary index for all the vehicles in the dataset. This is accomplished with the index_col parameter to the read_csv function
    df = pd.read_csv(file, index_col = 'vin')
    df.drop(columns=['Unnamed: 0'], inplace=True)

except FileNotFoundError:
  print(f'file not found in {WORKING_DIR}')
df.head(3)
```

```python
def multiply_cols(df: pd.DataFrame, col_1: str, col_2: str) -> pd.DataFrame:
    """
    Multiplies the values of two columns in a DataFrame and adds a new column with the results.

    Parameters:
    - df (pd.DataFrame): The input DataFrame.
    - col_1 (str): The name of the first column.
    - col_2 (str): The name of the second column.

    Returns:
    - pd.DataFrame: The DataFrame with an additional column containing the multiplied values.
    """

    new_col_name = f"{col_1}_{col_2}"
    df[new_col_name] = df[col_1] * df[col_2]
    return df


#calling this function on price and mileage

df_new = multiply_cols(df,col_1 = 'price',col_2 = 'mileage')

#result of multiply will be displayed in a new column called price_mileage
df_new.head(10)
```

### Task 7: Write a For loop in combination with Try/Except and the new_list. Modify the new_list as needed to test both paths of the Try/Except. Show evidence of the Except logic.

```python
#adding an integer to the list such it triggers the except block
car_brands_v2[9] = 2;

new_list = []

for index, item in enumerate(car_brands_v2):
    if item is None:
        print(f'there is a null value in the list on position {index}')
        new_list.append(item)
    else:
        try:
            upper_cased_item = item.upper()
            new_list.append(upper_cased_item)
        except AttributeError:
            print(f'invalid data type on {item}')

print(f'old list: {car_brands_v2}')
print(f'new list: {new_list}')
```

### Task 8: Create a dictionary with 10 key:value pairs (all lower case)

### Task 9: Use the dictionary created in the previous step to create a dateframe

```python
dataframe_rows = []
for rows in countries_presidents.items():
  dataframe_rows.append(rows)

#get a list of tuples which serve as rows in the dataframe
dataframe_rows
```

```python
df_presidents = pd.DataFrame(dataframe_rows,columns=['Country','President'])
df_presidents
```

### Task 10: Create a For loop for the dictionary created in the previous 2 steps. For each key, make it upper case and save the new key:value pair to the dictionary.  Print the dictionary to see the old and new key:value pairs.

```python
countries_presidents_copy = {}

# Iterate through the original dictionary and add uppercase keys to the copy
for key, value in countries_presidents.items():
    upper_key = key.upper()
    countries_presidents_copy[upper_key] = value

# Print the copied dictionary with both original and new key-value pairs
print(f'old dictionary: {countries_presidents}')
print(f'new dictionary: {countries_presidents_copy}')
```
